```python
data_final=df[df.title.str.contains('batman',case=False)]
```

```python
from sklearn.feature_extraction.text import CountVectorizer


countvec = CountVectorizer()

dtm = countvec.fit_transform(data_final['description'])
```

```python
dtm.toarray()
```

```python
print(dtm)
```

```python
from sklearn.metrics.pairwise import linear_kernel

cosine_sim_matrix = linear_kernel(dtm,dtm)

cosine_sim_matrix.shape
```

```python
data_final.reset_index(drop=True,inplace=True)
```

```python
titles = data_final['title']
```

```python
titles.shape
```

```python
indices = pd.Series(data_final.index,index=titles)
```

```python
indices
```

```python
def recommend(title):
  idx = indices[title]
  sim_scores = list(enumerate(cosine_sim_matrix[idx]))
  sim_scores = sorted(sim_scores, key = lambda x: x[1] , reverse=True)
  sim_scores = sim_scores[1:31]
  movie_indices = [i[0] for i in sim_scores]
  return titles.iloc[movie_indices]
```

```python
recommend('Batman Begins').head(10)
```
